## Corrélation entre la part de femmes entre 25 et 39 ans (15 à 29?) et la part de « Moins de 14 ans »

```python
# Corrélation entre la part de femmes entre 25 et 39 ans (15 à 29?) et la part de « Moins de 14 ans »

# Il s'agit de deux variables quantitatives

df_3 = df[['P18_F1529' , 'P18_POP0014']]
df_3
```

```python
# Représentation graphique des deux variables par un diagramme de dispersion (nuage de points):

plt.figure (figsize = (10,6))
sns.regplot(data=df_3, x = 'P18_F1529', y='P18_POP0014', fit_reg=False)
plt.show()
```

```python
# Notre question : Il y a t-il une correlation entre les variables?
# Les hypothèses:
    # H0 = il n'y a pas de correlation
    # H1 = il y a une correlation

# Comme il s'agit de deux variables quantitatives on va utiliser soit Pearson (paramétrique) soit Spearman (non paramétrique)
# Pour savoir quel test utiliser on doit tester nos variables par rapport à la loi de normalité
```

```python
# 1ère étape: Visualisation graphique de la distribution de la variable P18_F1529:

sns.histplot(data=df, x='P18_F1529', kde=True)
plt.show()
```

```python
# Est-ce que la variable 'P18_F1529' suit une distribution Gaussienne?
    # H0 = la variable suit une distribution Gaussienne
    # H1 = la variable ne suit pas une distribution Gaussienne
    
# Pour le savoir on utilise le test d'Anderson (shapiro est interessant pour une distribution < 5000)

st.anderson(df['P18_F1529'], dist='norm')

# Le résultat du test 'statistic' est ici > à la valeur qu'on a choisi dans 'critical_values' 
    #(0.787 pour un seuil alpha de 0.05) ==> on rejette l'hypothèse H0 : la variable ne suit pas une distribution Gaussienne
```

```python
# 2ème étape: Visualisation de la distribution de la variable price:

sns.histplot(data=df, x='P18_POP0014', kde=True)
plt.show()
```

```python
# Est-ce que la variable 'P18_POP0014' suit une distribution Gaussienne?
    # H0 = la variable suit une distribution Gaussienne
    # H1 = la variable ne suit pas une distribution Gaussienne
    
st.anderson(df['P18_POP0014'], dist = 'norm')

# Le résultat du test 'statistic' est ici > à la valeur qu'on a choisi dans 'critical_values' 
    #(0.787 pour un seuil alpha de 0.05) ==> on rejette l'hypothèse H0, la variable ne suit pas une distribution Gaussienne
```

```python
#Les deux variables ne suivent pas une distribution Gaussienne, on utilise donc le test de Spearman (non paramétrique):

st.spearmanr(df['P18_F1529'], df['P18_POP0014'])
```

```python
# Positive correlations imply that as x increases, so does y
```
